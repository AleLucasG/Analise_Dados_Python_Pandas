**Trabalhando com datas**

```python
#Importando a Biblioteca
import pandas as pd
```

```python
#Leitura dos Arquivos
df01 = pd.read_excel("Aracaju.xlsx")
df02 = pd.read_excel("Fortaleza.xlsx")
df03 = pd.read_excel("Natal.xlsx")
df04 = pd.read_excel("Recife.xlsx")
df05 = pd.read_excel("Salvador.xlsx")
```

```python
#unificação dos 5 arquivos
df = pd.concat([df01,df02,df03,df04,df05])
```

```python
df.head()
```

```python
#alteração do tipo de dado da coluna LojaId
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
#Criar coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#transformar coluna de data para interiro
df["Data"] = df["Data"].astype("int64")
```

```python
df.head()
```

```python
#verificando o tipo da coluna
df.dtypes
```

```python
# alterando o tipo de coluna inteiro para data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# valor da receita por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criar coluna ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#extraindo o mes e o dia da coluna Data
df["Mes_Venda"], df["Dia_Venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#vizualizando a data mais antida
df["Data"].min()
```

```python
# calculando a diferença de dias
df["Diferenca_Dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#criando coluna de trimestre
df["Trimestre_Vendas"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#filtrando vendas pelo ano d 2019
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
