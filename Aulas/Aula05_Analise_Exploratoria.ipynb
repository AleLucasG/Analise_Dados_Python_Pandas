```python
# Cria um novo DataFrame chamado "df" a partir de um arquivo Excel chamado "AdventureWorks.xlsx".
# É necessário que o arquivo esteja previamente carregado no ambiente (como no Google Colab, via upload).
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Tipos de dados
df.dtypes
```

```python
# Calcula a Receita Total somando todos os valores da coluna "Valor Venda".
df["Valor Venda"].sum()
```

```python
# Cria a coluna "Custo" calculando o custo total por produto, multiplicando o "Custo Unitário" pela "Quantidade"
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
# Cria a coluna "lucro" calculando a diferença entre o "Valor Venda" e o "Custo" para cada registro
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
# Exclui a coluna duplicada no DataFrame
df = df.drop(columns=["lucro"])
```

```python
df.head()
```

```python
# Retorna a soma da coluna "Custo" arredondada para 2 casas decimais
round(df["Custo"].sum(), 2)
```

```python
# Calcula o lucro total somando todos os valores da coluna "Lucro" e arredonda o resultado para 2 casas decimais
round(df["Lucro"].sum(), 2)
```

```python
# Cria uma nova coluna "Tempo_Envio" calculando a diferença entre as datas de envio ("Data Envio") e venda ("Data Venda").
# O resultado indica o tempo, em dias, que levou para o envio após a venda.
df["Tempo_Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

**Media de envio para cada Marca.**

```python
# Atualiza a coluna "Tempo_Envio" extraindo apenas o número de dias entre "Data Envio" e "Data Venda"
df["Tempo_Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
# Verifica o tipo de dado da coluna "Tempo_Envio"
df["Tempo_Envio"].dtypes
```

```python
# Calcula a média do tempo de envio para cada marca
df.groupby("Marca")["Tempo_Envio"].mean()
```

```python
# Calcula a média do tempo de envio para cada marca e converte o resultado para inteiro
df.groupby("Marca")["Tempo_Envio"].mean().round(0).astype(int)
```

```python
# Verifica se existem valores nulos em cada coluna e retorna a quantidade por coluna
df.isnull().sum()
```

```python
# Agrupa os dados por ano de venda e marca, somando os valores da coluna "Lucro" para cada grupo
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
# Agrupa os dados por ano de venda e marca, soma os valores da coluna "Lucro" e arredonda o resultado para 2 casas decimais
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().round(2)
```

```python
# Define o formato de exibição dos números float no Pandas para mostrar com duas casas decimais e separador de milhar
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
# Agrupa os dados por ano de venda e marca, soma os valores da coluna "Lucro",
# reseta o índice para transformar o MultiIndex em colunas normais e armazena no DataFrame "lucro_ano"
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
# Agrupa os dados por produto, soma a quantidade vendida de cada um e ordena do maior para o menor
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Gera um gráfico de barras horizontais mostrando o total de produtos vendidos, ordenados do menor para o maior
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto")
```

```python
# Gera um gráfico de barras mostrando o lucro total por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
```

```python
# Agrupa os dados por ano da venda e soma os valores da coluna "Lucro" para cada ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
# Filtra o DataFrame para trazer apenas as vendas do ano de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
# Agrupa os dados de 2009 por mês, soma o lucro de cada mês e gera um gráfico de linha (plot) com marcadores: marker="o"
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(marker="v")
plt.xlabel("Mês")
plt.ylabel("Lucro")
plt.title("Lucro x Mês")
```

```python
# Gera um gráfico de barras mostrando o lucro total por marca no ano de 2009
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```

```python
#Maior lucro por classe
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```

```python
# Exibe estatísticas descritivas (contagem, média, desvio, min, max, quartis) da coluna "Tempo_Envio"
df["Tempo_Envio"].describe()
```

```python
# Cria um gráfico boxplot para visualizar a distribuição do tempo de envio
plt.boxplot(df["Tempo_Envio"])
```

```python
import matplotlib.pyplot as plt

# Dados estatísticos
stats = {
    'count': 904,
    'mean': 8.54,
    'std': 3.06,
    'min': 4,
    '25%': 6,
    '50%': 9,
    '75%': 11,
    'max': 20
}

plt.boxplot(df["Tempo_Envio"])

# Adiciona texto com as estatísticas no gráfico
plt.text(1.1, stats['min'], f"Min: {stats['min']}", verticalalignment='bottom')
plt.text(1.1, stats['25%'], f"1º Quartil: {stats['25%']}", verticalalignment='bottom')
plt.text(1.1, stats['50%'], f"Mediana: {stats['50%']}", verticalalignment='bottom')
plt.text(1.1, stats['75%'], f"3º Quartil: {stats['75%']}", verticalalignment='bottom')
plt.text(1.1, stats['max'], f"Max: {stats['max']}", verticalalignment='bottom')
plt.text(1.1, stats['mean'], f"Média: {stats['mean']:.2f}", verticalalignment='bottom')
plt.text(1.1, stats['mean'] - stats['std'], f"Desvio Padrão: {stats['std']:.2f}", verticalalignment='bottom')

plt.title('Boxplot do Tempo de Envio')
plt.ylabel('Dias')

plt.show()
```
