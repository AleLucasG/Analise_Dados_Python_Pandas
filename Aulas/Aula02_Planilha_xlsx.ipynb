**Trabalhando com planilhas do excel**

```python
#Importando a Biblioteca
import pandas as pd
```

```python
#Leitura dos Arquivos
df01 = pd.read_excel("Aracaju.xlsx")
df02 = pd.read_excel("Fortaleza.xlsx")
df03 = pd.read_excel("Natal.xlsx")
df04 = pd.read_excel("Recife.xlsx")
df05 = pd.read_excel("Salvador.xlsx")
```

```python
df01.head()
```

```python
df02.head()
```

```python
#unificação dos 5 arquivos
df = pd.concat([df01,df02,df03,df04,df05])
```

```python
df.head()
```

```python
df.sample(5)
```

```python
#tipo de dado de cada coluna
df.dtypes
```

```python
#alteração do tipo de dado da coluna LojaId
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
#verifica valores null (se form null, me traga a soma)
df.isnull().sum()
```

```python
# Retorna uma amostra aleatória de 15 linhas do DataFrame "df"
# O método "sample(15)" seleciona 15 registros aleatórios do DataFrame para visualização ou análise.
df.sample(15)
```

```python
# Calcula a média dos valores na coluna "Vendas" do DataFrame "df"
# O método "mean()" retorna o valor médio de todos os registros da coluna "Vendas".
df["Vendas"].mean()
```

**Criando colunas novas**


```python
#Cria uma nova coluna "Receita"
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
# Retorna o valor da maior receita presente na coluna "Receita"
# O método "max()" encontra o maior valor dentro da coluna "Receita".
df["Receita"].max()
```

```python
# Retorna o valor da menor receita presente na coluna "Receita"
# O método "min()" encontra o menor valor dentro da coluna "Receita".
df["Receita"].min()
```

```python
# Retorna as 3 linhas com os maiores valores na coluna "Receita"
# O método "nlargest(3, 'Receita')" seleciona as 3 linhas com as maiores receitas.
df.nlargest(3, "Receita")
```

```python
# Retorna as 3 linhas com os menores valores na coluna "Receita"
# O método "nsmallest(3, 'Receita')" seleciona as 3 linhas com as menores receitas.
df.nsmallest(3, "Receita")
```

```python
# Agrupa os dados pelo valor da coluna "Cidade" e calcula a soma das receitas para cada cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
# Ordena os valores da coluna "Receita" em ordem decrescente (do maior para o menor)
# Em seguida, retorna as 10 primeiras linhas usando o método "head(10)".
df.sort_values("Receita", ascending=False).head(10)
```
